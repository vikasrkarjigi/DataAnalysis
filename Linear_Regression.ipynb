## **Linear Regression**

```python
import numpy as np
import matplotlib.pyplot as plt
#plt.style.use('./deeplearning.mplstyle')
```

```python
x_train = np.array([1.0, 2.0])
```

```python
y_train = np.array([300.0, 500.0])
```

```python
print(f"x_train = {x_train}")
print(f'y_train = {y_train}')
```

```python
print(f'x_train.shape : {x_train.shape}')
m = x_train.shape[0]
print(f'Number of training examples is: {m}')
m = len(x_train)
# Another way of finding the length of an array instead of using shape
print(f'Number of training examples is: {m}')
```

```python
for i in range(len(x_train)):
    print(f'(x^({i}), y^({i})) = ({x_train[i]}, {y_train[i]})')

print(f'Another method to print the ith value')

i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
plt.scatter(x_train, y_train, marker ='x', c ='r', label = 'Actual values')
plt.title('Housing Prices')
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
w = 200
b = 100
print (f'w: {w}\nb: {b}')
```

```python
def compute_model_output(x,w,b):
  m = x.shape[0]
  f_wb = np.zeros(m)
  for i in range(m):
    f_wb[i] = w*x[i] + b
  return f_wb
```

```python
temp_f_wb = compute_model_output(x_train, w, b)
print(f'temp_f_wb: {temp_f_wb}')
plt.plot(x_train, temp_f_wb, c='b', label = 'Our Prediction')
plt.scatter(x_train, y_train, marker = 'x', c ='r', label ='Actual Values')
plt.title("Housing Prices")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```
